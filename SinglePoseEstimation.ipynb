```python
import cv2 as cv
import matplotlib.pyplot as plt
import mediapipe as mp
```

```python
image_1 = cv.cvtColor(cv.imread('/content/drive/MyDrive/Colab Notebooks/human.jpg'), cv.COLOR_BGR2RGB)
plt.imshow(image_1)
```

```python
image_2 = cv.cvtColor(cv.imread('/content/drive/MyDrive/Colab Notebooks/humans.jpg'), cv.COLOR_BGR2RGB)
plt.imshow(image_2)
```

```python
print(image_1.shape)
print(image_2.shape)
```

```python
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

results = pose.process(image_1)
mpDraw.draw_landmarks(image_1, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
plt.imshow(image_1)
plt.show()

results = pose.process(image_2)
mpDraw.draw_landmarks(image_2, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
plt.imshow(image_2)
```
